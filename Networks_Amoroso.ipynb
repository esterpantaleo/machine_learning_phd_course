```python
import networkx as nx


N = 100
reps0 = 100

n_nodes = []
#see if a giant component appears or not, and find the size

for i in range(1,reps0-1):
    p = i/reps0
    G = nx.erdos_renyi_graph(N,p) #binomial model
    giant = len(max(nx.connected_components(G), key=len))
    n_nodes.append(giant)
    
n_nodes
ave_degree = p*N
```

```python
p=0.00001
G=nx.erdos_renyi_graph(N,p)
print(len(max(nx.connected_components(G), key=len)))
```
